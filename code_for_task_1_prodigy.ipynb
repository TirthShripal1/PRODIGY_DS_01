<h2>Importing libraries</h2>

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

<h2>Dataset </h2>


```python
url = 'https://github.com/AdithiVS/PRODIGY_DS_01/raw/main/worldpopulationdata.csv'
ds= pd.read_csv(url)
ds.head()
```

<h1>Info about the Data</h1>

```python
ds.info()
```

```python
ds.describe()
```

<h3>Checking for duplicate rows & Missing values</h3>

```python
ds.isna().sum()
```

```python
ds.duplicated().sum()
```

<h2>Dropping unnecessary cols</h2>

```python
ds.drop(['Series Name','Country Code'],axis=1,inplace=True)
ds.columns
```

<h2>Histogram of Total Population</h2>

```python
total_population_data = ds[ds['Series Code'] == 'SP.POP.TOTL'].copy()
numeric_columns = total_population_data.select_dtypes(include='number').columns
total_population_data['Row Total'] = total_population_data[numeric_columns].sum(axis=1)
plt.hist(total_population_data['Row Total'])
plt.xlabel('total population')
plt.ylabel('Frequency')
plt.title('Distribution of Total Population')
plt.show()
```

<h3>Extraction of top-10 countries with respect to total population</h3>

```python

total_population_sorted = total_population_data.sort_values(by="Row Total", ascending=False)
total_top_ten_countries = total_population_sorted.head(10)
print("Top ten countries with highest row totals of population total column\n")
print(total_top_ten_countries[['Country Name', 'Row Total']])
```

<h2>Bar plot</h2>

<h3>Top ten countries of total population</h3>

```python
# Create the bar plot
plt.figure(figsize=(15, 6))
plt.subplot(2,2,1)
sns.barplot(x="Row Total", y="Country Name", data=total_top_ten_countries, palette="plasma")
plt.title("Top Ten Countries of Total Population",fontsize=10)
plt.xlabel("Total Population",fontsize=10)
plt.ylabel("Country",fontsize=10)
plt.show()
```

<h2>Scatter plot</h2>

<h3>Top ten countries of total population</h3>

```python
plt.scatter(x="Row Total", y="Country Name", data=total_top_ten_countries,color='red')
plt.title("Scatter Plot",fontsize=10)
plt.xlabel("Total Population",fontsize=10)
plt.ylabel("Country",fontsize=10)
plt.show()
```
